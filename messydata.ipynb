```python
import pandas as pd

# Load the dataset
df = pd.read_csv('student_performance.csv')

# Clean column names
df.columns = df.columns.str.strip()

# 1. Remove duplicate rows
#  Purpose: To eliminate repeated entries that could bias analysis or model training
df = df.drop_duplicates()

# 2. Strip extra spaces from all string fields
#  Purpose: To ensure consistent formatting and avoid errors during filtering or mapping
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# 3. Convert numeric columns to proper types
#  Purpose: To enable mathematical operations, plotting, and model training
df['Hours_Studied'] = pd.to_numeric(df['Hours_Studied'], errors='coerce')
df['Attendance'] = pd.to_numeric(df['Attendance'], errors='coerce')
df['Final_Score'] = pd.to_numeric(df['Final_Score'], errors='coerce')

# 4. Standardize Result labels
#  Purpose: To unify values like 'pass', 'Pass', and ' PASS ' into a consistent format
df['Result'] = df['Result'].str.strip().str.capitalize()

# 5. Drop rows with missing or invalid data
#  Purpose: To ensure the dataset is clean and complete for modeling
df = df.dropna()

# Preview cleaned data
df.head()
```
